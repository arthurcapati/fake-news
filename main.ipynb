Explanability 

```python
mlflow.set_tracking_uri("http://127.0.0.1:8080/")
model_name = "fake-news"
model_version = "latest"

# Load the model from the Model Registry
model_uri = f"models:/{model_name}/{model_version}"
model = mlflow.sklearn.load_model(model_uri)
```

```python
dataprocessor = DataProcessor()
title1 = "Trump is an alien from other dimension"
pred1 = model.predict(dataprocessor.transform_features([title1]))
print(f"Titulo: {title1} - Prediction: {pred1}")

title2 = "Republicans press House leadership for help as they face pressure over DOGE cuts at home"
pred2 = model.predict(dataprocessor.transform_features([title2]))
print(f"Titulo: {title2} - Prediction: {pred2}")
```

```python
import shap

def explain(title):
    # Criar um modelo wrapper
    def model_predict(X):
        return model.predict(X)
    
    cv = DataProcessor().vectorizer
    background = np.zeros((1, cv.transform([""]).shape[1]))

    explainer = shap.KernelExplainer(model_predict, background)
    return explainer.shap_values(cv.transform([title]))

cv = DataProcessor().vectorizer
feature_names = cv.get_feature_names_out()  # Obt√©m as palavras do CountVectorizer
# Obt√©m os nomes das classes do modelo (Fake ou Real)
class_names = ["Fake", "Real"]  # Altere conforme necess√°rio para sua base de dados


title = title1
shap_values = explain(title)
shap.summary_plot(shap_values, cv.transform([title]), feature_names=feature_names, class_names=class_names, max_display=len(title.split(" ")))


title = title2
shap_values = explain(title)
shap.summary_plot(shap_values, cv.transform([title]), feature_names=feature_names, class_names=class_names, max_display=len(title.split(" ")))
```
